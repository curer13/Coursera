```python
import numpy as np
from scipy.stats import expon, norm
import matplotlib.pyplot as plt
from statsmodels.distributions.empirical_distribution import ECDF
```

```python
lambda_ = 1.5
exp = expon(loc=0, scale=1/lambda_)
sample = exp.rvs(1000)
x = np.linspace(0, 5, 100)
```

```python
plt.hist(sample, bins=30, density=True);
plt.plot(x, exp.pdf(x), label='pdf')
plt.xlabel('$x$');
plt.ylabel('fraction of samples');
plt.legend(loc='upper right');
```

```python
def CLT(n):
    matrix = exp.rvs(1000).reshape(1000,1)
    for feature in range(n):
        col = exp.rvs(1000).reshape(1000,1)
        matrix = np.hstack([matrix, col])
    return matrix.mean(axis=1)
```

```python
# дисперсия 1/lambda**2
# матожидание 1/lambda
DX = 1/lambda_**2
EX = 1/lambda_
sigma = np.sqrt(DX)
DX, EX
```

```python
N = [5, 10, 50, 80]
# fig, ax = plt.subplots(1, 4, figsize=(20, 4));
i = 1

for n in N:
    sample = CLT(n)
    s_sample = sorted(sample)
    # plt.subplot(1, 4, i);
    plt.xlim([0, 2]);
    plt.ylim([0, 6])
    plt.hist(sample, bins=30, density=True, label=f'n={n}');
    plt.plot(s_sample, norm.pdf(s_sample, loc=np.mean(s_sample), scale=np.std(s_sample)));
    plt.xlabel('$x$');
    plt.ylabel('fraction of samples');
    # plt.legend(loc='upper right');
    plt.title(f'n={n}');
    plt.show();
    print(f'Теоритическая дисперсия нормального распределения:{DX/n}')
    print(f'Эмпирическая дисперсия нормального распределения:{np.std(s_sample)**2}')
    i += 1
```
